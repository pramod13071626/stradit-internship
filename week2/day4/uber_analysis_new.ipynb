```python
df_features = df_completed.copy()

is_weekend = df_features['Date'].dt.dayofweek >= 5
df_features['is_weekend'] = is_weekend.astype(int)
df_features['Is_Weekend'] = df_features['is_weekend']

print("\nFeature engineering complete. 'is_weekend' feature added to df_features dataset.")
```

```python
# Required imports
from scipy.stats import (
    ttest_ind,
    mannwhitneyu,
    levene,
    shapiro
)

# Completed rides only
df_hypo = df_features[
    df_features['Booking Status'] == 'Completed'
].copy()

# Ensure numeric datatype
df_hypo['Avg CTAT'] = pd.to_numeric(
    df_hypo['Avg CTAT'],
    errors='coerce'
)

# Remove nulls
df_hypo = df_hypo.dropna(subset=['Avg CTAT'])

# Create groups
weekday_duration = df_hypo[
    df_hypo['is_weekend'] == 0
]['Avg CTAT']

weekend_duration = df_hypo[
    df_hypo['is_weekend'] == 1
]['Avg CTAT']

print("\nDATA SUMMARY")
print("-" * 50)

print(f"Weekday rides: {len(weekday_duration)}")
print(f"Weekend rides: {len(weekend_duration)}")

print(f"\nWeekday Mean Duration: {weekday_duration.mean():.2f} min")
print(f"Weekend Mean Duration: {weekend_duration.mean():.2f} min")

# ============================================================================
# STEP 1: HYPOTHESIS DEFINITION
# ============================================================================

print("\nHYPOTHESIS SETUP")
print("-" * 50)

print("Null Hypothesis (H0):")
print("There is NO significant difference in ride duration")
print("between weekdays and weekends.")

print("\nAlternative Hypothesis (H1):")
print("There IS a significant difference in ride duration")
print("between weekdays and weekends.")

alpha = 0.05

print(f"\nSignificance Level (alpha): {alpha}")


# ============================================================================
# STEP 3: VARIANCE HOMOGENEITY TEST
# ============================================================================

print("\nVARIANCE EQUALITY TEST (Levene Test)")
print("-" * 50)

levene_test = levene(
    weekday_duration,
    weekend_duration
)

print(f"Levene Statistic: {levene_test.statistic:.4f}")
print(f"p-value: {levene_test.pvalue:.5f}")

equal_variance = levene_test.pvalue > alpha

if equal_variance:
    print("Variances are equal")
else:
    print("Variances are NOT equal")

# ============================================================================
# STEP 4: INDEPENDENT T-TEST
# ============================================================================

print("\nINDEPENDENT T-TEST")
print("-" * 50)

t_test = ttest_ind(
    weekday_duration,
    weekend_duration,
    equal_var=equal_variance
)

print(f"T-Statistic: {t_test.statistic:.4f}")
print(f"P-Value: {t_test.pvalue:.6f}")

# Decision
if t_test.pvalue < alpha:

    print("\nRESULT: Reject Null Hypothesis")

    print(
        "There IS a statistically significant difference "
        "between weekday and weekend ride durations."
    )

else:

    print("\nRESULT: Fail to Reject Null Hypothesis")

    print(
        "There is NO statistically significant difference "
        "between weekday and weekend ride durations."
    )


# ============================================================================
# STEP 7: CONFIDENCE INTERVAL
# ============================================================================

print("\n95% CONFIDENCE INTERVAL")
print("-" * 50)

mean_weekday = weekday_duration.mean()
mean_weekend = weekend_duration.mean()

diff_mean = mean_weekend - mean_weekday

se_diff = np.sqrt(

    (
        weekday_duration.var() / len(weekday_duration)
    )

    +

    (
        weekend_duration.var() / len(weekend_duration)
    )
)

margin_error = 1.96 * se_diff

ci_lower = diff_mean - margin_error
ci_upper = diff_mean + margin_error

print(
    f"Mean Difference (Weekend - Weekday): "
    f"{diff_mean:.2f} min"
)

print(
    f"95% CI: "
    f"[{ci_lower:.2f}, {ci_upper:.2f}]"
)

# ============================================================================
# FINAL BUSINESS INTERPRETATION
# ============================================================================

print("\n" + "=" * 80)
print("BUSINESS INTERPRETATION")
print("=" * 80)

if t_test.pvalue < alpha:

    if diff_mean > 0:

        print(
            f"\nWeekend rides are significantly LONGER "
            f"than weekday rides by approximately "
            f"{abs(diff_mean):.2f} minutes."
        )

        print("\nPossible Reasons:")
        print("- Leisure and long-distance trips")
        print("- Weekend traffic congestion")
        print("- Airport / outstation travel")

    else:

        print(
            f"\nWeekend rides are significantly SHORTER "
            f"than weekday rides by approximately "
            f"{abs(diff_mean):.2f} minutes."
        )

        print("\nPossible Reasons:")
        print("- Lower office traffic")
        print("- Faster road movement")
        print("- More local trips")

else:

    print(
        "\nRide durations are operationally stable "
        "across weekdays and weekends."
    )

print("\nStatistical hypothesis testing complete!")
```
