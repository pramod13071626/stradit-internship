```python
# 7. Matrix Operations Perform: Matrix transpose Matrix multiplication Determinant Inverse on a 3x3 matrix.

matrix = np.array([[1, 2, 3],
                   [0, 1, 4],
                     [5, 6, 0]])

# Transpose
transpose = np.transpose(matrix)
print("Transpose of the matrix:\n", transpose)

# Matrix Multiplication (matrix multiplied by its transpose)
matrix_multiplication = np.dot(matrix, transpose)
print("Matrix multiplication (matrix * transpose):\n", matrix_multiplication)

# Determinant
determinant = np.linalg.det(matrix)
print("Determinant of the matrix:", determinant)

# Inverse
inverse = np.linalg.inv(matrix)
print("Inverse of the matrix:\n", inverse)

```

```python
# 8. Missing Value Handling Replace all NaN values in array with: Mean of array

array_with_nan = np.array([1, 2, np.nan, 4, 5])
mean_value = np.nanmean(array_with_nan)
array_without_nan = np.where(np.isnan(array_with_nan), mean_value, array_with_nan)
print("Array with NaN values:", array_with_nan)
print("Array after replacing NaN with mean:", array_without_nan)
```
