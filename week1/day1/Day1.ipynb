```python
# Program to find even and odd

num=int(input("Enter a number"))
if num % 2 == 0:                           # we can also use num % 2 != 0 for odd
        print("Even")
else:
    print("odd")
```

```python
num=int(input("Enter a number"))
if num & 1:                                   # I used the bitwise operator to check for odd or even using the last bit of the number.
        print("Odd")
else:
    print("Even")
```

```python
# Write a program to find the largest among three numbers.
a=float(input("Enter the first Number:"))
b=float(input("Enter the Second Number:"))
c=float(input("Enter the third Number:"))

""" we have to compare all the three numbers with each other.
    we will compare each one number with other"""

if a>=b and a>=c:            #basic comparison of numbers using if-else statements.
    print("Greatest:",a)
elif b>=a and b>=c:
    print("Greatest:",b)
else:
    print("Greatest:",c)
    
```

```python
# Write a Python program to calculate factorial of a number.
def fact(n):
    if n==1 or n==0:
        return 1
    return n * fact(n-1)               #It helps to calculate the factorial of a number using recursion.

num = int(input("Enter a number:"))

print("Factorial is:",fact(num))
```

```python
# Write a program to print the Fibonacci series up to N terms.
n=int(input("Enter the number"))
a=0
b=1
print("Fibonacci Series")

for i in range(n):
    print(a,end="")  # we will print the initial value of a and the last value of b.
    c=a+b
    a=b
    b=c
```

```python
n=int(input("Enter the Number:"))
a,b=0,1
print("Finonacci Series")

for i in range(n):
    print(a,end="")

    a,b=b,a+b       # a,b is assinged the values of b and a+b each.
```

```python
# Reverse a string without using built-in reverse functions.
text=input("Enter the text=")
l=list(text)
for  i in range(len(l)-1,-1,-1):    # we interate the list in reverse order and print each character using the -1 logic.
    print(l[i], end="") 
```

```python
# Write a program to count vowels and consonants in a string.

text="Pramod"
v= "aeiouAEIOU"             # we assigned all the vowels to the v variable.
vowels, consonants =0,0     # at start we will have the count initialized to 0.
for char in text:           # we now iterate throughthe text for every character.
    if char in v:
        vowels+=1           # if the character is vowel then we will increase the count of vowels by 1.
    else:
        consonants+=1       # if the character is not vowel then we will increase the count of consonants by 1.

print(f"Vowels: {vowels}, Consonants: {consonants}")
```

```python
text="Pramod"
v= "aeiouAEIOU"
vowels, consonants =0,0
vowels = sum(1 for char in text if char in v)
consonants = sum(1 for char in text if char not in v)

print(f"Vowels: {vowels}, Consonants: {consonants}")
```

```python
# Write a Python program to check whether a string is palindrome or not.'''

text=input("Enter the text=")

if text== text[::-1]:            # we used the slicing method to reverse the string and check if its a palindrome or not.
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
"""Initailly we will reverse the string using the slicing method and then compare it with the original string to check if its a palindrome or not.
For that we created the rev function then reverse the string and then the the function is called to check if the string is palindrome or not."""
def rev(text):
    reverse=""

    for i in range(len(text)-1,-1,-1):
        reverse=reverse + text[i]

    return reverse
text=input("Enter the text=")
reverse=rev(text)
print("Original text:",text)
print("Reverse:",reverse)

if text==reverse:
    print("Palindrome")
else:
    print("Not palindrome")
```
