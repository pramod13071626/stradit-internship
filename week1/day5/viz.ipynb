```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# 24. Titanic Dataset Analysis

# Using Titanic dataset visualize:
# Survival count

# Gender distribution

# Age distribution

# Passenger class distribution



csv_file_path = 'C:\\Users\\Lenovo\\Documents\\DE\\Week1\\DAY 1\\stradit-internship\\week1\\day5\\train.csv'

titanic_df = pd.read_csv(csv_file_path)

print(titanic_df.head())

# Survival count

survival_count = titanic_df['Survived'].value_counts()
plt.figure(figsize=(6, 4))
survival_count.plot(kind='bar', color=['red', 'green'])
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Gender distribution

gender_distribution = titanic_df['Sex'].value_counts()
plt.figure(figsize=(6, 4 ))
gender_distribution.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Age distribution

plt.figure(figsize=(8, 5))
plt.hist(titanic_df['Age'].dropna(), bins=20, color='lightblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Passenger class distribution

pclass_distribution = titanic_df['Pclass'].value_counts()
plt.figure(figsize=(6, 4))
pclass_distribution.plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

    
```

```python
# 25. Employee Analytics Dashboard

# Create visualizations for:
# Salary distribution

# Department-wise average salary

# Employee count by department

# Experience vs salary trend

csv_file_path = 'C:\\Users\\Lenovo\\Documents\\DE\\Week1\\DAY 1\\stradit-internship\\week1\\day5\\employees.csv'

emp_df = pd.read_csv(csv_file_path)

print(emp_df.head())


# Salary distribution

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.hist(emp_df['Salary'], bins=5, color='lightblue', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(axis='y')

# Department-wise average salary

avg_salary_by_department = emp_df.groupby('Department')['Salary'].mean()
avg_salary_by_department.plot(kind='bar', color='orange')
plt.title('Department-wise Average Salary')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Employee count by department

department_counts = emp_df['Department'].value_counts()
department_counts.plot(kind='bar', color='lightgreen')
plt.title('Employee Count by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Experience vs salary trend

plt.subplot(2, 2, 4)
plt.scatter(emp_df  ['Experience'], emp_df['Salary'], color='purple', edgecolor='black')
plt.title('Experience vs Salary Trend')

plt.xlabel('Experience (years)')
plt.ylabel('Salary')
plt.grid()
plt.tight_layout()
plt.show()  


```

```python
# 23. Sales Dashboard

# Using sales dataset create:
# Monthly sales trend

# Top-selling products

# Region-wise sales chart

csv_file_path = 'C:\\Users\\Lenovo\\Documents\\DE\\Week1\\DAY 1\\stradit-internship\\week1\\day5\\sales_data.csv'

sales_df = pd.read_csv(csv_file_path)

sales_df['Date'] = pd.to_datetime(sales_df['Date'])
sales_df['Month'] = sales_df['Date'].dt.to_period('M')

monthly_sales_trend = sales_df.groupby('Month')['Sales'].sum()

top_selling_products = sales_df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(5)

region_sales = sales_df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
monthly_sales_trend.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()

plt.subplot(2, 2, 2)


top_selling_products.plot(kind='bar', color='orange')

plt.title('Top-Selling Products')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()

plt.subplot(2, 2, 3)
region_sales.plot(kind='bar', color='lightgreen')
plt.title('Region-wise Sales')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()


```

```python
# Problem 1 — Student Marks Bar Chart

# You are given marks of students in Python.
# Create a bar chart showing:
# student names
# marks obtained

# Requirements:
# Add chart title
# Add axis labels
# Display marks on top of bars

# Example Data:
# students = ['Rahul', 'Priya', 'Amit', 'Sara']
# marks = [85, 92, 78, 88]

students = ['Rahul', 'Priya', 'Amit', 'Sara']
marks = [85, 92, 78, 88]

plt.figure(figsize=(8, 5))
plt.bar(students, marks, color='skyblue')
plt.title('Marks Obtained by Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.ylim(0, 100)
plt.grid(axis='y')
for index, value in enumerate(marks):
    plt.text(index, value + 1, str(value), ha='center')
plt.tight_layout()
plt.show()


```

```python


# Problem 2 — Monthly Sales Line Chart

# A company wants to analyze monthly sales trends.
# Create a line chart showing:
# months on X-axis
# sales on Y-axis

# Requirements:
# Use markers
# Add grid
# Add legend

# Example Data:
# months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
# sales = [10000, 15000, 12000, 18000, 25000]


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [10000, 15000, 12000, 18000, 25000]

plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', color='blue', label='Sales')
plt.title('Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()


# Problem 3 — Product Share Pie Chart

# Visualize market share of products using a pie chart.

# Requirements:
# Show percentage
# Highlight highest-selling product
# Add title

# Example Data:
# products = ['Mobile', 'Laptop', 'Tablet', 'Watch']
# sales = [45, 30, 15, 10]


products = ['Mobile', 'Laptop', 'Tablet', 'Watch']
sales = [45, 30, 15, 10]
plt.figure(figsize=(8, 8))
plt.pie(sales, labels=products, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral', 'lightgreen', 'lightyellow'])

plt.title('Market Share of Products')
plt.show()


# Problem 4 — Temperature Trend Visualization

# Create a line chart to visualize weekly temperature changes.

# Requirements:
# Highlight maximum temperature
# Highlight minimum temperature
# Use different colors for markers

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 24, 19, 27, 25, 30, 28]
plt.figure(figsize=(8, 5))
plt.plot(days, temperatures, marker='o', color='blue', label='Temperature')
plt.title('Weekly Temperature Trend')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()



# Problem 5 — Employee Salary Histogram

# Visualize salary distribution using histogram.

# Requirements:
# Create salary bins
# Show frequency distribution
# Explain which salary range has most employees


salaries = [30000, 45000, 50000, 60000, 75000, 80000, 90000, 100000, 120000, 150000]
plt.figure(figsize=(8, 5))
plt.hist(salaries, bins=5, color='lightblue', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```
