```python
# DICTIONARY TASKS

# 12. Create a dictionary for employee details: name department salary Print all keys and values.

employee = {
    "name": "John Doe",
    "department": "Sales",
    "salary": 50000
}

print("Employee details:")
for key, value in employee.items():   # We are traversing through the dictionary.

    print(f"{key}: {value}")



# 13. Count frequency of characters in a string using dictionary. Example: "Pramod Deore"

string = "Pramod Deore"
freq= {}

for char in string:
    if char in freq:
        freq[char] += 1      # If a character is present in string we are increasing the count of frequency by 1.
    else:
        freq[char] = 1
print("Character frequency:")

for char, count in freq.items():   # 
    print(f"{char}: {count}")


# 14. Store student marks in dictionary and: Find topper Find average marks

marks = {
    "pramod": 85,
    "Sai": 90,
    "Vaibhav": 78,
    "Nikhil": 92
}

# Find topper
top = max(marks, key=marks.get)                 # Here we are using the key with max function.

print(f"Topper: {top} with {marks[top]} marks")

# Find average marks
average= sum(marks.values()) / len(marks)    # we are aggregating the marks by getting marks with values method.

print(f"Average marks: {average}")



# 15. Merge two dictionaries into one.

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "d": 4}
merged_dict = {**dict1, **dict2} # The ** operator unpacks a dictionary into key-value pairs. 
                                # the value of key "b" comes from dict2, because when keys overlap, the last unpacked dictionary overrides earlier ones.

print("Merged dictionary:", merged_dict)
```

```python
# LIST COMPREHENSION TASKS

# 16. Create a list of squares from 1 to 10 using list comprehension.

sqr = [x**2 for x in range(1, 11)]     # list comprehension is used here.
print("Squares from 1 to 10:", sqr)


# 17. Create a list of even numbers from 1 to 50 using list comprehension.

even = [num for num in range(1, 51) if num % 2 == 0]     # list comprehension is used here.

print("Even numbers from 1 to 50:", even)

# 18. Convert all names to uppercase using list comprehension.

names = ["Pramod", "Vijay", "Sanjay"]
names = [name.upper() for name in names]            # Python provides us .upper method to convert string to uppercase.

print("Names in uppercase:", names)


# 19. Remove empty strings from a list using list comprehension.

strings = ["Pramod", "", "is", "", "data Engineer"]
empty = [s for s in strings if s]    # String is false in boolean context if empty.

print("Non-empty strings:", empty)


# 20. Create a list of lengths of words.

words = ["Pramod", "Vijay", "Sanjay"]
lengths = [len(word) for word in words]   
print("Lengths of words:", lengths)


```

```python
# DICTIONARY COMPREHENSION TASKS

# 21. Create a dictionary of numbers and their squares.

squares = {x: x**2 for x in range(1, 11)}     # dict comprehension same as list comprehension.
print("Dictionary of numbers and their squares:", squares)


# 22. Convert temperature list from Celsius to Fahrenheit using dictionary comprehension.

cel= [0, 10, 20, 30, 40]
fah = {c: (c * 9/5) + 32 for c in cel}              # we used the formula to convert cel to far.
print("Temperatures in Fahrenheit:", fah)


# 23. Create dictionary from two lists: keys list values list

keys = ["name", "age", "city"]
values = ["Pramod", 24, "Nashik"] 


dict = {k: v for k, v in zip(keys, values)}     # we are getting the key value pairs and then use the zip function.
print("Dictionary from two lists:", dict)




```

```python
# SET COMPREHENSION TASKS

# 24. Create a set of unique vowels from a string.

string = "Pramod Deore"

vowels = {char for char in string if char.lower() in 'aeiou'}
print("Unique vowels in the string:", vowels)


# 25. Create a set of squares of even numbers from 1 to 20.

sqr_even = {x**2 for x in range(1, 21) if x % 2 == 0}           # Modulo operator used to identify even and then set comprehension is done.

print("Squares of even numbers from 1 to 20:", sqr_even)


```

```python
# MIXED PRACTICE TASKS


#26. Given a list of numbers: Create list of squares Create list of cubes
#  Create dictionary of number:square 
# Create set of even numbers using comprehensions only.


num = [1, 2, 3, 4, 5]

sqr = [x**2 for x in num]

cube = [x**3 for x in num]

sqr_d = {x: x**2 for x in num}

even = {x for x in num if x % 2 == 0}

print("Squares:", sqr)

print("Cubes:", cube)

print("Dictionary of number:square:", sqr_d)

print("Set of even numbers:", even)


# 27. Given employee data: employees = [
    # {"name":"Rahul","salary":50000},
    # {"name":"Aman","salary":60000},
    #{"name":"Priya","salary":55000}]
   

# Tasks: Create list of employee names Create list of salaries above 55000 Create dictionary of name:salary using comprehensions.

employees = [
    {"name": "Rahul", "salary": 50000}, 
    {"name": "Aman", "salary": 60000}, 
    {"name": "Priya", "salary": 55000}]

n= [emp["name"] for emp in employees]                                        # We are traversing through the list of dictionary and getting the name of employee and creating the list of names.
sal_55000 = [emp["salary"] for emp in employees if emp["salary"] > 55000]    # We are getting the emp whoose salary > 55000.
dict = {emp["name"]: emp["salary"] for emp in employees}

print("Employee names:", n)
print("Salaries above 55000:", sal_55000)   
print("Dictionary of name:salary:", dict)


# 29. Create a nested dictionary for students and print: Highest marks Lowest marks Average marks

students = {
    "Pramod": {"marks": 85},
    "Vaishnavi": {"marks": 90},
    "Ankita": {"marks": 78},
    "Ekta": {"marks": 92}
}

marks = [student["marks"] for student in students.values()]             # we create a list from nested dict to get marks of students.
high = max(marks)   
low = min(marks)                                                           # Using the inbuilt functions.
ave = sum(marks) / len(marks)

print("Highest marks:", high)
print("Lowest marks:", low)
print("Average marks:", ave)

```
